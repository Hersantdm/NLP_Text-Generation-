```python
# mengimport dataframe berisi transrkrip dengan tanda baca
# import packages
import requests
from bs4 import BeautifulSoup
import pickle

# membuat fungsi url to transcript
def url_to_transcript(url):
    page = requests.get(url).text
    soup = BeautifulSoup(page,'lxml')
    text = [p.text for p in soup.find_all('div', class_ = 'elementor-widget-theme-post-content')]
    print(url)
    return text
urls = ['https://scrapsfromtheloft.com/movies/spider-man-no-way-home-transcript/',
        'https://scrapsfromtheloft.com/movies/eternals-2021-transcript/',
        'https://scrapsfromtheloft.com/comedy/venom-let-there-be-carnage-transcript/',
        'https://scrapsfromtheloft.com/comedy/shang-chi-legend-of-ten-rings-transcript/',
        'https://scrapsfromtheloft.com/movies/zack-snyders-justice-league-transcript/',
        'https://scrapsfromtheloft.com/movies/joker-how-about-another-joke-murray-transcript/',
        'https://scrapsfromtheloft.com/movies/wonder-woman-1984-transcript/']
movies = ['spiderman', 'eternals', 'venom', 
          'shang-chi','justice league','joker',
          'wonder woman']
transcripts = [url_to_transcript(i) for i in urls]

# create & load data
for i, c in enumerate(movies):
    with open('transcripts/' + c + '.txt', 'wb') as file:
        pickle.dump(transcripts[i],file)
data = {}
for i, c in enumerate(movies):
    with open("transcripts/" + c + ".txt", "rb") as file:
        data[c] = pickle.load(file)
        

# menggabungkan text menjadi satu bagian
def combine_text(list_of_text):
    combined_text = ' '.join(list_of_text)
    return combined_text
data_combined = {key: [combine_text(value)] for (key,value) in data.items()}

# membuat dataframe data_df
import pandas as pd
pd.set_option('max_colwidth',150)

data_df = pd.DataFrame.from_dict(data_combined).transpose()
data_df.columns = ['transcript']
data_df = data_df.sort_index()
```

```python
# data cleaning (menghapus new line)
import re

def clean_text_round2(text):
    text = re.sub('[‘’“”…♪]', '', text)
    text = re.sub('\n', ' ', text)
    text = re.sub('\*', ' ', text)
    return text

round2 = lambda x: clean_text_round2(x)

data_clean = pd.DataFrame(data_df.transcript.apply(round2))
```

# Membuat Markov Chain

```python
from collections import defaultdict

def markov_chain(text):
    # tokenization
    words = text.split(' ')
    # initalize dictionary baru
    m_dict = defaultdict(list)
    # membuat list berisi kata dengan kata berikutnya
    for current_word, next_word in zip(words[0:-1], words[1:]):
        m_dict[current_word].append(next_word)
        
    m_dict = dict(m_dict)
    return m_dict
```

# Text Generation Function

```python
import random

def generate_sentence(chain, count):
    # Huruf pertama kita set kapital
    word1 = random.choice(list(chain.keys()))
    sentence = word1.capitalize()
    
    # men-generate kata berikutnya
    for i in range(count-1):
        word_n = random.choice(chain[word1])
        word1 = word_n
        sentence += ' ' + word_n
        
    # tutup kalimat dengan titik
    sentence += '.'
    
    return(sentence)
```

```python
for i in range (10):
    print(generate_sentence(spd_chain, 15), end = ' ') 
```

```python
# Untuk film venom
vnm_text = data_clean.transcript.loc['venom']
vnm_chain = markov_chain(vnm_text)
for i in range (10):
    print(generate_sentence(vnm_chain, 15), end = ' ') 
```

```python
# coba membuat dari data tanpa tanda baca
import string 

def clean_text_round20(text):
    text = re.sub('[‘’“”…♪]', '', text)
    text = re.sub('\n', ' ', text)
    text = re.sub('\*', ' ', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = text.lower()
    return text

round20 = lambda x: clean_text_round20(x)

data_clean20 = pd.DataFrame(data_df.transcript.apply(round20))
```

```python
# Untuk film venom
vnm_text = data_clean20.transcript.loc['venom']
vnm_chain = markov_chain(vnm_text)
for i in range (10):
    print(generate_sentence(vnm_chain, 15), end = ' ') 
```
