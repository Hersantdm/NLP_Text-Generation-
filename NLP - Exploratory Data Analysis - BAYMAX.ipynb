# Top Words

```python
import pandas as pd

data = pd.read_pickle('dtm.pkl')
data = data.transpose()
data.head()
```

```python
# mencari 30 kata yang paling sering muncul di masing-masing film
top_dict = {}
for c in data.columns:
    top = data[c].sort_values(ascending = False).head(30)
    top_dict[c] = list(zip(top.index, top.values))
```

```python
top_dict
```

```python
for movies, top_words in top_dict.items():
    print(movies)
    print(', '.join([word for word, count in top_words[0:30]]))
    print('---')
```

```python
from collections import Counter

# meng-ekstrak 30 kata terbanyak di masing-masing film
words = []
for movie in data.columns:
    top = [word for (word,count) in top_dict[movie]]
    for t in top:
        words.append(t)
words
```

```python
Counter(words).most_common()
```

```python
# tambah kata2 yang paling sering muncul di semua film ke list stop_words
add_stop_words = [word for word, count in Counter(words).most_common() if count > 3]
add_stop_words
```

```python
from sklearn.feature_extraction import text
from sklearn.feature_extraction.text import CountVectorizer

data_clean = pd.read_pickle('data_clean.pkl')
```

```python
# menggabung kata sambung yg kita temukan dengan yg sudah tersedia di library
stop_words = text.ENGLISH_STOP_WORDS.union(add_stop_words)

# recreate DTM
cv = CountVectorizer(stop_words = stop_words)
data_cv = cv.fit_transform(data_clean.transcript)
data_stop = pd.DataFrame(data_cv.toarray(), columns = cv.get_feature_names())
data_stop.index = data_clean.index

# pickle untuk digunakan pada pertemuan berikutnya
import pickle
pickle.dump(cv, open('cv_stop.pkl','wb'))
data_stop.to_pickle('dtm_stop.pkl')
```

```python
# membuat wordcloud
from wordcloud import WordCloud

wc1 = WordCloud(stopwords = stop_words, colormap='Dark2', background_color='white'
               , max_font_size = 150, random_state = 42)
```

```python
data.columns
```

```python
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [16,6]

for index, movie in enumerate(data.columns):
    wc1.generate(data_clean.transcript[movie])
    plt.subplot(3, 4, index+1)
    plt.imshow(wc1, interpolation = 'bilinear')
    plt.axis('off')
    plt.title(data.columns[index])

plt.show()
```

insights:
1. paling sering muncul nama karakternya di masing2 film

# Number of Unique Words

```python
movies = data.columns
```

```python
# Identify the non-zero items in the document-term matrix, meaning that the word occurs at least once
unique_list = []
for movie in data.columns:
    uniques = data[movie].to_numpy().nonzero()[0].size
    unique_list.append(uniques)

# Create a new dataframe that contains this unique word count
data_words = pd.DataFrame(list(zip(movies, unique_list)), columns=['movie', 'unique_words'])
data_unique_sort = data_words.sort_values(by='unique_words')
data_unique_sort
```

```python
# mencari total words
total_list = []
for movie in data.columns:
    totals = sum(data[movie])
    total_list.append(totals)
duration = [132, 10, 242, 148, 151, 97, 122]
```

```python
data_words['total_words'] = total_list
data_words['duration'] = duration
data_words['words_per_minute'] = data_words['total_words'] / data_words['duration']
data_words
```

```python
data_unique_sort = data_words.sort_values(by='unique_words')
data_unique_sort
```

```python
data_wpm_sort = data_words.sort_values(by='words_per_minute')
data_wpm_sort
```

Insight:
1. berdasarkan unique words, spiderman mirip dengan wonder woman dan eternals
2. berdasarkan wpm, spiderman mirip wonder woman dan venom
